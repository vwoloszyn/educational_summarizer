```python
import pandas as pd
dataset=pd.read_csv("./data/movies_twm.csv", encoding = "ISO-8859-1")

from bs4 import BeautifulSoup

import requests

    
    
```

```python
def between(left,right,s):
    before,_,a = s.partition(left)
    a,_,after = a.partition(right)
    return a    
    
    
def get_id_amazon_from_twm_url(url):
    r  = requests.get(url)

    data = r.text
    soup = BeautifulSoup(data)
    print (url)
    for link in soup.find_all('a'):
        str_link = str(link.get('href'))
        if ( "www.amazon.com" in str_link):
            print (str_link)
            id_amazon=between("/product/","/ref=",str_link)
            print (id_amazon)
            return id_amazon
            break

dataset['id_amazon']= dataset.apply(lambda x: get_id_amazon_from_twm_url(x['url']),axis=1 )
```
